## Part 2


```python
from sklearn.preprocessing import StandardScaler

# 1️⃣ Выбираем числовые признаки (все кроме ID)
num_features = df_active.select_dtypes(include=['number']).columns.drop('ID', errors='ignore')

# 2️⃣ Масштабируем данные
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_active[num_features])

print("Shape of feature matrix:", X_scaled.shape)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Проверим несколько k
inertias = []
for k in range(2, 11):
    km = KMeans(n_clusters=k, n_init=20, random_state=42)
    km.fit(X_scaled)
    inertias.append(km.inertia_)

plt.plot(range(2, 11), inertias, marker='o')
plt.title('KMeans Elbow Method')
plt.xlabel('Number of clusters k')
plt.ylabel('Inertia')
plt.show()

# Оптимальный k (например, 4–6)
best_k = 5
kmeans = KMeans(n_clusters=best_k, n_init=20, random_state=42)
df_active['cluster_km'] = kmeans.fit_predict(X_scaled)
```

```python
from sklearn.cluster import AgglomerativeClustering

agg = AgglomerativeClustering(n_clusters=best_k, linkage='ward')
df_active['cluster_ag'] = agg.fit_predict(X_scaled)
```

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

def evaluate(X, labels):
    return {
        "Silhouette": silhouette_score(X, labels),
        "Calinski-Harabasz": calinski_harabasz_score(X, labels),
        "Davies-Bouldin": davies_bouldin_score(X, labels)
    }

print("KMeans:", evaluate(X_scaled, df_active['cluster_km']))
print("Agglomerative:", evaluate(X_scaled, df_active['cluster_ag']))
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(6,5))
plt.scatter(X_pca[:,0], X_pca[:,1], c=df_active['cluster_km'], cmap='viridis', s=10)
plt.title('KMeans Clusters (PCA projection)')
plt.xlabel('PC1'); plt.ylabel('PC2')
plt.show()
```

Segmentation methods.
We applied two clustering algorithms to the standardized dataset of 3 665 active clients (34 features): K-Means and Agglomerative (Ward linkage).
Using the Elbow method, the optimal number of clusters was estimated around k = 5.
K-Means achieved a higher Silhouette score (0.146 vs 0.131) and a lower Davies–Bouldin index (1.87 vs 2.13) than Agglomerative, indicating better internal consistency.
PCA visualization confirms five distinct groups with moderate overlap.
Therefore, K-Means was selected as the primary segmentation method and Agglomerative as a complementary approach.

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import numpy as np
import pandas as pd

def eval_labels(X, labels):
    if len(set(labels)) < 2:
        return dict(sil=np.nan, ch=np.nan, db=np.nan)
    return dict(
        sil = silhouette_score(X, labels),
        ch  = calinski_harabasz_score(X, labels),
        db  = davies_bouldin_score(X, labels)
    )

rows = []
bic_rows, aic_rows = [], []
K = range(3, 11)

for k in K:
    # KMeans
    km = KMeans(n_clusters=k, n_init=20, random_state=42).fit(X_scaled)
    m = eval_labels(X_scaled, km.labels_)
    rows.append(dict(method='KMeans', k=k, **m, inertia=km.inertia_))

    # Agglomerative (Ward)
    ag = AgglomerativeClustering(n_clusters=k, linkage='ward')
    ag_lab = ag.fit_predict(X_scaled)
    m = eval_labels(X_scaled, ag_lab)
    rows.append(dict(method='Agglomerative', k=k, **m, inertia=np.nan))

    # GMM
    gmm = GaussianMixture(n_components=k, covariance_type='full', random_state=42).fit(X_scaled)
    gmm_lab = gmm.predict(X_scaled)
    m = eval_labels(X_scaled, gmm_lab)
    rows.append(dict(method='GMM', k=k, **m, inertia=np.nan))
    bic_rows.append(dict(k=k, bic=gmm.bic(X_scaled)))
    aic_rows.append(dict(k=k, aic=gmm.aic(X_scaled)))

metrics_df = pd.DataFrame(rows).sort_values(['method','k'])
bic_df = pd.DataFrame(bic_rows).sort_values('k')
aic_df = pd.DataFrame(aic_rows).sort_values('k')

display(metrics_df)
display(bic_df)
display(aic_df)

# ---- выберем финальные модели (пример: k=5) ----
BEST_K = 5
kmeans = KMeans(n_clusters=BEST_K, n_init=20, random_state=42).fit(X_scaled)
df_active['cluster_km'] = kmeans.labels_

gmm_best = GaussianMixture(n_components=BEST_K, covariance_type='full', random_state=42).fit(X_scaled)
df_active['cluster_gmm'] = gmm_best.predict(X_scaled)

# сравнение согласованности сегментаций
from sklearn.metrics import adjusted_rand_score
print("ARI(KMeans vs GMM):", adjusted_rand_score(df_active['cluster_km'], df_active['cluster_gmm']))
```

Results of segmentation evaluation.
We compared K-Means, Agglomerative, and Gaussian Mixture Models for k = 3–10.
K-Means provided the best balance of compactness and separation with a Silhouette = 0.146 and a Davies–Bouldin = 1.87 at k = 5.
Agglomerative clustering achieved slightly lower quality (Sil ≈ 0.13), while GMM reached the highest Silhouette at k = 3 (0.19) but required more components (k ≈ 8–9) to minimize BIC/AIC.
Based on these metrics, we selected K-Means (k = 5) as the primary segmentation method and GMM (k = 3–5) as a supporting approach for profile comparison in Part 4.
The adjusted Rand index (ARI = 0.33) confirms moderate agreement between the two methods.

```python
import matplotlib.pyplot as plt

subset = metrics_df[metrics_df['method'] == 'KMeans']
plt.plot(subset['k'], subset['sil'], marker='o')
plt.title('Silhouette by k (KMeans)')
plt.xlabel('k'); plt.ylabel('Silhouette'); plt.show()

plt.plot(bic_df['k'], bic_df['bic'], marker='o', label='BIC')
plt.plot(aic_df['k'], aic_df['aic'], marker='o', label='AIC')
plt.legend(); plt.title('GMM: Information Criteria'); plt.xlabel('k'); plt.ylabel('Score'); plt.show()
```

## Part 4


```python
# Числовые и категориальные признаки
num_cols = df_active.select_dtypes(include=['number']).columns
cat_cols = df_active.select_dtypes(include=['object','category']).columns

# --- Функция для профилирования ---
def profile_clusters(df, cluster_col, num_cols, cat_cols):
    num_summary = df.groupby(cluster_col)[num_cols].mean().round(2)
    cat_summary = (
        df.groupby(cluster_col)[cat_cols]
          .agg(lambda x: x.value_counts().index[0])  # most frequent category
    )
    return num_summary, cat_summary

# KMeans
km_num, km_cat = profile_clusters(df_active, 'cluster_km', num_cols, cat_cols)

# GMM
gmm_num, gmm_cat = profile_clusters(df_active, 'cluster_gmm', num_cols, cat_cols)

# Просмотр
print("KMeans — numeric profile:")
display(km_num)
print("KMeans — categorical profile:")
display(km_cat)

print("GMM — numeric profile:")
display(gmm_num)
print("GMM — categorical profile:")
display(gmm_cat)
```

```python
overall = km_num.mean()
for cid, row in km_num.iterrows():
    diffs = (row - overall).sort_values(ascending=False)
    top_up = diffs.head(3).index.tolist()
    top_down = diffs.tail(3).index.tolist()
    print(f"Cluster {cid}: higher in {', '.join(top_up)}, lower in {', '.join(top_down)}.")
```

Overall, the segmentation demonstrates a robust and interpretable structure that can support targeted marketing, credit scoring refinement, and customer retention strategies.
